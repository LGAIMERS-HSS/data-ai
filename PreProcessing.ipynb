```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv("./Data/train.csv")
train_data = train_data.iloc[:,2:]
train_data
```

# 결측치 비율이 80% 이상인 columns 제거
* 나머지 column 제거는 직접 이름 넣어서 제거해도 괜찮을 것 같당

```python
train_data = train_data.loc[:,train_data.count() / len(train_data) > 0.2]
train_data
```

# 특정 시술 유형 Unknown, 결측치 처리

```python
len(train_data.loc[train_data["특정 시술 유형"].isna()])
```

```python
# 결측치 처리
train_data = train_data.dropna(subset=['특정 시술 유형'], axis='index')
len(train_data.loc[train_data["특정 시술 유형"].isna()])
```

```python
# Unknown 처리
# 특정 시술 유형이 Unknown인 경우, 시술 유형으로 대체
train_unknown = train_data.loc[train_data["특정 시술 유형"] == "Unknown"]
train_data.loc[train_unknown.index, "특정 시술 유형"] = train_data.loc[train_unknown.index, "시술 유형"]

print(train_data["특정 시술 유형"].value_counts())
for val in train_data["특정 시술 유형"].unique():
    if "Unknown" in val: # 특정 시술 유형의 범주에 Unknown 혼합형이 들어가면 Unknown만 무시
        train_data.loc[train_data["특정 시술 유형"]==val,"특정 시술 유형"] = val.split(":Unknown")[0]
    elif val == "ICSI / BLASTOCYST :IVF / BLASTOCYST": # 띄어쓰기 차이로 다른 범주로 처리되는 범주들 통일
        train_data.loc[train_data["특정 시술 유형"]==val,"특정 시술 유형"] = 'ICSI / BLASTOCYST:IVF / BLASTOCYST'

print(train_data["특정 시술 유형"].value_counts())
```

# 시술 유형이 DI인 경우
* 배아, 미세주입에 관한 column에 결측치가 있으면 0으로 채우기
* 시술 유형이 DI가 아닌데, 배아, 미세주입에 대한 column에 결측치가 있으면 행을 제거

```python
train_data
```

```python
# 시술 유형이 DI인 데이터 찾기
train_DI = train_data.loc[train_data["시술 유형"]=="DI"]
train_IVF = train_data.loc[train_data["시술 유형"]=="IVF"]
print(len(train_IVF))

print(train_data.loc[:,train_data.columns.str.contains("배아") ].count())

# 시술 유형이 DI이면서 배아 관련 열인 경우 결측치를 0으로 대체
train_data.loc[train_DI.index, train_data.columns.str.contains("배아")] = train_data.loc[train_DI.index, 
train_data.columns.str.contains("배아")].fillna(0)

print()

# 시술 유형이 IVF이면서 배아 관련 열인 경우 결측치 행을 제거
train_data = train_data.dropna(subset=['배아 이식 경과일'], axis='index')
print(train_data.loc[:,train_data.columns.str.contains("배아") ].count())
```

```python
# 배아 관련이 아니지만 DI 관련인 열에 대해서도 결측치 처리
# 특정 열 리스트
columns_fill = [
    '미세주입된 난자 수', '해동 난자 수', '수집된 신선 난자 수', 
    '파트너 정자와 혼합된 난자 수', '기증자 정자와 혼합된 난자 수', 
    '저장된 신선 난자 수', '혼합된 난자 수', '착상 전 유전 진단 사용 여부', '대리모 여부'
]

print(train_data.loc[:,columns_fill].count())

# '시술 유형'이 "DI"인 행에 대해 결측치 0으로 채우기
train_data.loc[train_DI.index, columns_fill] = train_data.loc[train_DI.index, columns_fill
].fillna(0)

print()

print(train_data.loc[:,columns_fill].count())
```

# 결과에 영향 없는 columns 제거
- 난자 채취 경과일: 0 아니면 결측치라 의미 없다고 판단
- 난자 혼합 경과일: 값이 달라져도 결과 차이 x

```python
train_data = train_data.drop(['난자 채취 경과일', '난자 혼합 경과일'], axis='columns')
```
