# 이상치 처리
- 가장 큰 값이 있는 행만 제거

```python
# 수치형 컬럼, 범주형 컬럼 구분
train_columns = train_data.columns.tolist()

categorical_columns = []
numerical_columns = []

for index, row in specification_data.iterrows():
    column_name = row['컬럼명']
    if column_name in train_columns:  
        if row['범주형 여부'] == 1:  # 범주형이면
            categorical_columns.append(column_name)
        elif row['범주형 여부'] == 0:  # 수치형이면
            numerical_columns.append(column_name)
```

```python
# 이상치 처리 (임신 성공 여부 0일 때만)
def remove_top_n_outliers(train_data, columns, n=1):
    for column in columns:
        if column in train_data.columns:
            sorted_values = train_data[column].sort_values(ascending=False)
            top_n_values = sorted_values.unique()[:n] 
            train_data = train_data[~((train_data['임신 성공 여부'] == 0) & (train_data[column].isin(top_n_values)))]
    return train_data

columns_to_check1 = ['미세주입된 난자 수', '저장된 배아 수', '미세주입 후 저장된 배아 수', '해동 난자 수', '저장된 신선 난자 수', '파트너 정자와 혼합된 난자 수']  # 최댓값 1개만 제거할 컬럼들
columns_to_check2 = ['해동된 배아 수', '혼합된 난자 수', '기증자 정자와 혼합된 난자 수'] # 최댓값 2개 제거할 컬럼들

train_data = remove_top_n_outliers(train_data, columns_to_check1, n=1)
train_data = remove_top_n_outliers(train_data, columns_to_check2, n=2)
```

# 데이터 변환
- 정규화/표준화 적용

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# 수치형 컬럼 업데이트
numerical_columns = [col for col in numerical_columns if col in train_data.columns]

# 정규화 (Min-Max Scaling)
scaler_minmax = MinMaxScaler()
train_norm = train_data.copy()
train_norm[numerical_columns] = scaler_minmax.fit_transform(train_norm[numerical_columns])

# 표준화 (Standardization)
scaler_standard = StandardScaler()
train_stand = train_data.copy()
train_stand[numerical_columns] = scaler_standard.fit_transform(train_stand[numerical_columns])
```

# 인코딩
현재는 정규화된 데이터로 진행
- 나이, 횟수 관련 컬럼 -> 라벨인코딩
- 순서 상관없는 나머지 컬럼 -> 원핫인코딩

```python
# 데이터는 정규화된 데이터로 지정
train_select = train_norm.copy()

# 범주형 데이터 문자열로 바꾸기
train_select[categorical_columns] = train_select[categorical_columns].astype(str)

# 인코딩별 컬럼 나누기
label_columns = ['시술 당시 나이', '배아 생성 주요 이유', '총 시술 횟수', '클리닉 내 총 시술 횟수', 'IVF 시술 횟수', 'DI 시술 횟수', 
                 '총 임신 횟수', 'IVF 임신 횟수', 'DI 임신 횟수', '총 출산 횟수','IVF 출산 횟수', 'DI 출산 횟수', '난자 기증자 나이', '정자 기증자 나이']

onehot_columns = [col for col in categorical_columns if col not in label_columns]
```

```python
# 라벨 인코딩
from sklearn.preprocessing import LabelEncoder

for col in label_columns:
    encoder = LabelEncoder()    
    train_select[col] = encoder.fit_transform(train_select[col])

# 확인
train_select[label_columns].head()
```

```python
#  원핫 인코딩
train_select = pd.get_dummies(train_select, columns = onehot_columns)
print(train_select.columns)
```
